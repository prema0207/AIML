```python
import cv2
import numpy as np

def detect_obstacle(frame):
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    # Define color range for obstacles (e.g., red object as obstacle)
    lower_red = np.array([0, 120, 70])
    upper_red = np.array([10, 255, 255])
    mask1 = cv2.inRange(hsv, lower_red, upper_red)

    # Detect another range of red
    lower_red2 = np.array([170, 120, 70])
    upper_red2 = np.array([180, 255, 255])
    mask2 = cv2.inRange(hsv, lower_red2, upper_red2)

    # Combine masks
    mask = mask1 + mask2
    output = cv2.bitwise_and(frame, frame, mask=mask)

    contours, _ = cv2.findContours(mask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    obstacle_detected = False

    for cnt in contours:
        area = cv2.contourArea(cnt)
        if area > 500:
            obstacle_detected = True
            x, y, w, h = cv2.boundingRect(cnt)
            cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 0, 255), 2)
            cv2.putText(frame, 'Obstacle Detected!', (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)

    return frame, obstacle_detected

# Use webcam or replace with image/video for simulation
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    processed_frame, obstacle = detect_obstacle(frame)

    if obstacle:
        cv2.putText(processed_frame, 'ACTION: Stop or Reroute', (50, 50),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 255), 3)
    else:
        cv2.putText(processed_frame, 'ACTION: Move Forward', (50, 50),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 3)

    cv2.imshow("Autonomous Navigation", processed_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

def make_decision(frame):
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    # Define red color as obstacle
    lower_red1 = np.array([0, 120, 70])
    upper_red1 = np.array([10, 255, 255])
    lower_red2 = np.array([170, 120, 70])
    upper_red2 = np.array([180, 255, 255])

    # Create masks
    mask1 = cv2.inRange(hsv, lower_red1, upper_red1)
    mask2 = cv2.inRange(hsv, lower_red2, upper_red2)
    mask = mask1 + mask2

    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    decision = "MOVE FORWARD"

    for cnt in contours:
        area = cv2.contourArea(cnt)
        if area > 500:
            x, y, w, h = cv2.boundingRect(cnt)
            cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 0, 255), 2)
            decision = "STOP - Obstacle Ahead"
            cv2.putText(frame, "Obstacle", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 255), 2)

    # Show decision on screen
    if decision == "MOVE FORWARD":
        color = (0, 255, 0)
    else:
        color = (0, 255, 255)

    cv2.putText(frame, f"ACTION: {decision}", (20, 40), cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 2)
    return frame

# Read input image simulating sensor input
frame = cv2.imread('test_navigation.jpg')  # Replace with your image
if frame is not None:
    processed = make_decision(frame)
    cv2.imshow("Autonomous Navigation Output", processed)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
else:
    print("Image not found. Please check the file path.")
```
